```python
import plotly.express as px
import pandas as pd
```

#Data Preparation

```python
url = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'
data_world = pd.read_csv(url,',')
data_GER = data_world['location'] == 'Germany'
data_IND = data_world['location'] == 'India'
data_USA = data_world['location'] == 'United States'

data_my_list = data_GER|data_IND|data_USA
data_vis = data_world[data_my_list]
```

#Relative Covid Cases Visualization

```python
data_vis['total_cases_norm'] = data_vis['total_cases']/data_vis['population']

fig = px.line(data_vis, x='date', y='total_cases_norm', color='location', title='Covid Cases')
fig.show()
```

#Vaccination Rate Visualization

```python
data_vis['vaccination_rate'] = data_vis['people_fully_vaccinated']/data_vis['population']*100

fig = px.line(data_vis, x='date', y='vaccination_rate', color='location', title='Vaccination Rate')
fig.show()
```
